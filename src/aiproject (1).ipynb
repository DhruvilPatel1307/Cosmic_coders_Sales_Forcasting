```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset = 'C:\\Users\\Nira\\Downloads\\archive (1)\\dairy_dataset.csv'
df = pd.read_csv(dataset)
```

```python
def total_revenue_by_product(df):
    plt.figure(figsize=(10, 6))
    sns.barplot(x='Product Name', y='Approx. Total Revenue(INR)', data=df, hue='Product Name')  # Added hue
    plt.title('Total Revenue by Product')
    plt.xlabel('Product Name')
    plt.ylabel('Total Revenue (INR)')
    plt.xticks(rotation=45)
    plt.show()
```

```python
def quantity_vs_price(df):
    plt.figure(figsize=(8, 6))
    sns.scatterplot(x='Quantity Sold (liters/kg)', y='Price per Unit (sold)', data=df, hue='Product Name', palette='Set1')
    plt.title('Quantity Sold vs. Price per Unit')
    plt.xlabel('Quantity Sold (liters/kg)')
    plt.ylabel('Price per Unit (sold)')
    plt.show()
```

```python
def stock_levels_vs_threshold(df):
    plt.figure(figsize=(10, 6))
    df['Stock Deficit'] = df['Minimum Stock Threshold (liters/kg)'] - df['Quantity in Stock (liters/kg)']
    sns.barplot(x='Product Name', y='Stock Deficit', data=df, hue='Product Name')  # Added hue
    plt.title('Stock Levels vs Minimum Stock Threshold')
    plt.xlabel('Product Name')
    plt.ylabel('Stock Deficit (liters/kg)')
    plt.xticks(rotation=45)
    plt.show()
```

```python
def shelf_life_distribution(df):
    plt.figure(figsize=(8, 6))
    sns.histplot(df['Shelf Life (days)'], bins=10, kde=True, color='purple')
    plt.title('Shelf Life Distribution of Products')
    plt.xlabel('Shelf Life (days)')
    plt.ylabel('Frequency')
    plt.show()
```

```python
def total_revenue_by_location(df):
    plt.figure(figsize=(10, 6))
    revenue_by_location = df.groupby('Location')['Approx. Total Revenue(INR)'].sum().reset_index()
    sns.barplot(x='Location', y='Approx. Total Revenue(INR)', data=revenue_by_location, hue='Location')  # Added hue
    plt.title('Total Revenue by Location')
    plt.xlabel('Location')
    plt.ylabel('Total Revenue (INR)')
    plt.xticks(rotation=45)
    plt.show()
```

```python
def sales_channel_distribution(df):
    plt.figure(figsize=(8, 6))
    sales_channel_counts = df['Sales Channel'].value_counts()
    sales_channel_counts.plot(kind='pie', autopct='%1.1f%%', colors=['lightblue', 'lightgreen'], startangle=90, legend=False)
    plt.title('Sales Channel Distribution')
    plt.ylabel('')  # Hide the 'y' label
    plt.show()
```

```python

if __name__ == "__main__":
    # Generate all the plots
    total_revenue_by_product(df)
    quantity_vs_price(df)
    stock_levels_vs_threshold(df)
    shelf_life_distribution(df)
    total_revenue_by_location(df)
    sales_channel_distribution(df)
```
